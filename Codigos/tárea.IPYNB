# Demostrar el centro de masa de un triángulo cuyos vértices son ABC tienen como vectores posición a,b y c 
# pespectivamente y relativos a un origen común O. Demuestre que el vector posición g del centróide G del 
# triángulo viene dado por: 
# g = (a+b+c)/3

Ahora, debido a que los vectores de posición a,b,c tienen un mismo origen común, es necesario definir los vectores que son 
los lados del triángulo. 
Para ello definimos: 

# BA = OA - OB
# BC = OC - OB
# AC = OC - OA 

Es decir, decimos que OA = a, OB = b, OC = c.

Para hallar el centróide del triángulo se requiere: 
1. Hallar los puntos medios de cada lado 
2. Trazar las rectas que unen los vértices con el punto medio del lado contrario. 

En este caso, siendo un problema vectorial (Los vectores previamente definidos están todos en notación vectorial), escalamos los vectores de los lados por:
# a = 1/2

Es decir

# BM = aBC = 1/2 * BC = 1/2 * (OC - OB)
# BP = aBA = 1/2 * BA = 1/2 * (OA - OB)
# AN = aAC = 1/2 * AC = 1/2 * (OC - OA)

Y las medianas son (lo queremos demostrar) concurrentes en un mismo punto que llamaremos G. 

Para esto, partimos de definir los puntos medios, entonces: 

# BC = m
# BA = p
# AC = n

Siendo m,p,n los puntos medios. 

Decimos que entonces los vectores del punto medio son:

# BG = alphaBN (BN es el vector que va desde el vértice B hasta el punto medio de AC y así con todos)
# CG = uCP 
# AG = rAM

Y aquí es trabajar vectorialmente para hallar las operaciones necesarias para despejar

# BG = alphaBN = alpha(BC + CN) = alpha(c-b-0.5(c-a)) = alpha(c/2 + a/2 - b)
# CG = uCP = u(CA + AP) = U(a - c -0.5(a-b)) = u(a/2 + b/2 - c)
# AG = rAM = r(AB + BM) = r(b - a +1/2 (c-b)) = r(a/2 + b/2 -c)

Recordamos que por las propiedades del centroide, alpha, u, y r deben ser iguales. 

También 

# AG = AB + BG = b - a + alpha(c/2 + a/2 - b) = b - a + alphac/2 + alphaa/2 - alphab/2
